```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import plotly.express as px
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Importing datasets

# Loading the data for study 1 and looking at head
df1 = pd.read_csv('data/data_study1.csv')

# Setting option to display unlimited number of columns
pd.set_option('display.max_columns', None)

df1.head()
```

```python
# Loading the data from the second study

df2 = pd.read_csv('data/data_study2.csv')

df2.head()
```

```python
# Dropping old index columns

df1.drop(columns='Unnamed: 0', axis=1, inplace=True)

# Sanity check
df1.head()
```

```python
df2.drop(columns='Unnamed: 0', axis=1, inplace=True)

# Sanity check
df2.head()
```

```python
# Creating a simple version of df2 for baseline modelling

df2_baseline = df2.drop(columns=['Weights',
                                 'CHES_version',
                                 'zlrgen',
                                 'zlrecon',
                                 'zgaltan',
                                 'ZPO',
                                 'ZmeanPO',
                                 'CPO',
                                 'CMQ_1',
                                 'CMQ_2',
                                 'CMQ_3',
                                 'CMQ_4',
                                 'CMQ_5',
                                 'CM4x'])

# Sanity Check

df2_baseline.head()
```

```python
# Checking the head

df2_baseline.info()
```

```python
# Converting sex column to binary

df2_baseline['Sex'] = np.where(df2_baseline['Sex'] == 'Male', 0, 1)

# Sanity Check

df2_baseline.head()
```

```python
# Sanity Check 2

df2_baseline['Sex'].unique()
```

```python
# Dropping NA values for baseline model

df2_baseline.dropna(inplace=True)

df2_baseline.info()
```

```python
# Creating dummy variables for country

df2_baseline = pd.get_dummies(df2_baseline, drop_first=True)

df2_baseline.head()
```

```python
# Defining X and y for Baseline Model

X = df2_baseline.drop(columns='CM5x')
y = df2_baseline['CM5x']

print(X.shape)
print(y.shape)
```

```python
# Train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, stratify=y, random_state=0)
```

```python
# Scaling X_train and X_test

my_scaler = StandardScaler()
X_train_scaled = my_scaler.fit_transform(X_train)
X_test_scaled = my_scaler.transform(X_test)
```

```python
# Trying baseline linear regression with scaled data

# Instantiate model
linmod = LinearRegression()

# Fitting to the scaled data
linmod.fit(X_train_scaled, y_train)

# Scoring on original train and test sets
print(f'R2 for Train Data: {linmod.score(X_train_scaled, y_train)}')
print(f'R2 for Test Data: {linmod.score(X_test_scaled, y_test)}')
```

```python
# Displaying the coefficients

coef_dict = {}
for coef, feature in list(zip(linmod.coef_, X.columns)):
    coef_dict[feature] = coef

coef_df = pd.DataFrame(coef_dict, index=range(1))
coef_df.T
```

```python
# Converting CM5x to binary column for baseline logistic regression model

# Converting CM5x to 10 point scale

y = y-1

y.head()
```

```python
# Calculate standard deviation
sd = stats.tstd(y)
mean = y.mean()
one_sd_from_mean = mean+sd
print(f"The standard deviation for CM5x is {sd}")
print(f"The mean for CM5x is {mean}")
print(f"The mean plus one standard deviation is {one_sd_from_mean}")
```

```python
# Showing distribution of y using histogram
plt.figure(figsize=(10,7))

plt.hist(y, color='skyblue')
plt.title('Histogram of Average CMQ Scores in the Second Dataset')
plt.axvline(x=mean+sd, color='r', linestyle='--', label='1 Standard deviation from mean (7.9)')
plt.axvline(x=mean, color='green', linestyle='dotted', label='Mean (5.8)')
plt.xlabel('Average CMQ Score')
plt.ylabel('Number of Observations')
plt.legend()
sns.despine()
plt.show()
```

```python
# Converting y_train and y_test CM5x to high (> 1 SD from mean) and not high (<= 1 SD from mean)
y_train = np.where(y_train > one_sd_from_mean, 1, 0)
y_test = np.where(y_test > one_sd_from_mean, 1, 0)

# Sanity Check
print(y_train[:5])
print(y_test[:5])
```

```python
# Trying a logistic regression with our new target

# Instantiate and fit
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)

# Scoring on train and test
print(f'Train Score: {logreg.score(X_train_scaled, y_train)}')
print(f'Test Score: {logreg.score(X_test_scaled, y_test)}')
```

```python
# Comparing with KNN

# Trying KNN with Ks in range from 1 to 20

Ks = range(1,20)

train_acc = []
test_acc = []

for K in Ks:
    KNN_model = KNeighborsClassifier(n_neighbors=K)
    KNN_model.fit(X_train_scaled, y_train)

    # Score the model on the test set
    train_predictions = KNN_model.predict(X_train_scaled)
    test_predictions = KNN_model.predict(X_test_scaled)
    train_accuracy = accuracy_score(train_predictions, y_train)
    test_accuracy = accuracy_score(test_predictions, y_test)
    train_acc.append(train_accuracy)
    test_acc.append(test_accuracy)
```

```python
# Plotting train vs test for different K values

plt.figure()
plt.plot(Ks, train_acc, color='red', label='Train')
plt.plot(Ks, test_acc, color='blue', label='Test')
plt.xticks(Ks)
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.title("Train and Test Accuracy for KNN Models With Different K Values")
plt.legend()
sns.despine()
plt.show()
```
