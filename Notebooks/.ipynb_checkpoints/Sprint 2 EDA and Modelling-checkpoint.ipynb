```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import plotly.express as px
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score, precision_score, recall_score
```

```python
# Importing datasets

# Loading the data for study 1 and looking at head
df1 = pd.read_csv('data/data_study1.csv')

# Setting option to display unlimited number of columns
pd.set_option('display.max_columns', None)

df1.head()
```

```python
# Loading the data from the second study

df2 = pd.read_csv('data/data_study2.csv')

df2.head()
```

```python
# Dropping old index columns

df1.drop(columns='Unnamed: 0', axis=1, inplace=True)

# Sanity check
df1.head()
```

```python
df2.drop(columns='Unnamed: 0', axis=1, inplace=True)

# Sanity check
df2.head()
```

```python
# Creating a simple version of df2 for baseline modelling

df2_baseline = df2.drop(columns=['Weights',
                                 'CHES_version',
                                 'zlrgen',
                                 'zlrecon',
                                 'zgaltan',
                                 'ZPO',
                                 'ZmeanPO',
                                 'CPO',
                                 'CMQ_1',
                                 'CMQ_2',
                                 'CMQ_3',
                                 'CMQ_4',
                                 'CMQ_5',
                                 'CM4x'])

# Sanity Check

df2_baseline.head()
```

```python
# Checking the head

df2_baseline.info()
```

```python
# Converting sex column to binary

df2_baseline['Sex'] = np.where(df2_baseline['Sex'] == 'Male', 0, 1)

# Sanity Check

df2_baseline.head()
```

```python
# Sanity Check 2

df2_baseline['Sex'].unique()
```

```python
# Dropping NA values for baseline model

df2_baseline.dropna(inplace=True)

df2_baseline.info()
```

```python
# Creating dummy variables for country

df2_baseline = pd.get_dummies(df2_baseline, drop_first=True)

df2_baseline.head()
```

```python
# Defining X and y for Baseline Model

X = df2_baseline.drop(columns='CM5x')
y = df2_baseline['CM5x']

print(X.shape)
print(y.shape)
```

```python
# Train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, stratify=y, random_state=0)
```

```python
# Scaling X_train and X_test

my_scaler = StandardScaler()
X_train_scaled = my_scaler.fit_transform(X_train)
X_test_scaled = my_scaler.transform(X_test)
```

```python
# Trying baseline linear regression with scaled data

# Instantiate model
linmod = LinearRegression()

# Fitting to the scaled data
linmod.fit(X_train_scaled, y_train)

# Scoring on original train and test sets
print(f'R2 for Train Data: {linmod.score(X_train_scaled, y_train)}')
print(f'R2 for Test Data: {linmod.score(X_test_scaled, y_test)}')
```

```python
# Displaying the coefficients

coef_dict = {}
for coef, feature in list(zip(linmod.coef_, X.columns)):
    coef_dict[feature] = coef

coef_df = pd.DataFrame(coef_dict, index=range(1))
coef_df.T.sort_values(by=[0], ascending=False)
```

```python
# Converting CM5x to binary column for baseline logistic regression model

# Converting CM5x to 10 point scale

y = y-1

y.head()
```

```python
# Calculate standard deviation
sd = stats.tstd(y)
mean = y.mean()
one_sd_from_mean = mean+sd
print(f"The standard deviation for CM5x is {sd}")
print(f"The mean for CM5x is {mean}")
print(f"The mean plus one standard deviation is {one_sd_from_mean}")
```

```python
# Showing distribution of y using histogram
plt.figure(figsize=(10,7))

plt.hist(y, color='skyblue', alpha=.7)
plt.title('Histogram of Average CMQ Scores in the Second Dataset')
plt.axvline(x=mean+sd, color='salmon', linestyle='dotted', label='1 Standard deviation from mean (7.9)')
plt.axvline(x=mean, color='grey', linestyle='--', label='Mean (5.8)')
plt.xlabel('Average CMQ Score')
plt.ylabel('Number of Observations')
plt.legend()
sns.despine()
plt.show()
```

```python
# Converting y_train and y_test CM5x to high (> 1 SD from mean) and not high (<= 1 SD from mean)
y_train = np.where(y_train > mean, 1, 0)
y_test = np.where(y_test > mean, 1, 0)

# Sanity Check
print(y_train[:5])
print(y_test[:5])
```

```python
# Trying a logistic regression with our new target

# Instantiate and fit
logreg = LogisticRegression(class_weight={0:1,1:1})
logreg.fit(X_train_scaled, y_train)

# Scoring on train and test
print(f'Train Score: {logreg.score(X_train_scaled, y_train)}')
print(f'Test Score: {logreg.score(X_test_scaled, y_test)}')
```

```python
# Generating confusion matrix for logistic regression model, train

ConfusionMatrixDisplay.from_estimator(logreg, X_train_scaled, y_train);

y_train_pred = logreg.predict(X_train_scaled)

print("====================================================")
print("Model Evaluation and Confusion Matrix for Train Data")
print("====================================================\n")
print(classification_report(y_train, y_train_pred))
```

```python
# Generating confusion matrix for logistic regression model, test

ConfusionMatrixDisplay.from_estimator(logreg, X_test_scaled, y_test);

y_test_pred = logreg.predict(X_test_scaled)

print("===================================================")
print("Model Evaluation and Confusion Matrix for Test Data")
print("===================================================\n")
print(classification_report(y_test, y_test_pred))
```

```python
# Plotting ROC Curve for Train

from sklearn.metrics import RocCurveDisplay

plt.figure()

plt.subplots(1,2, figsize=(15,5))

plt.subplot(1,2,1)
RocCurveDisplay.from_estimator(logreg, X_train_scaled, y_train, ax=plt.gca(), color='skyblue')
plt.title('ROC Curve for Train Data')

plt.subplot(1,2,2)
RocCurveDisplay.from_estimator(logreg, X_test_scaled, y_test, ax=plt.gca(), color='salmon')
plt.title('ROC Curve for Test Data')

plt.tight_layout
sns.despine()
plt.show()
```

```python
# Comparing Accuracy, Recall, and Precision for Models with different weights

train_accs = []
train_precs = []
train_recs = []

test_accs = []
test_precs = []
test_recs = []

for w in np.arange(1,11,.5):
    
    # Instantiating and fitting model
    logreg = LogisticRegression(class_weight={0:1,1:w})
    logreg.fit(X_train_scaled, y_train)
    
    # Creating variables for predictions
    y_train_pred = logreg.predict(X_train_scaled)
    y_test_pred = logreg.predict(X_test_scaled)
    
    # Recording accuracy,precision and recall for train and test
    train_accs.append(logreg.score(X_train_scaled, y_train))
    test_accs.append(logreg.score(X_test_scaled, y_test))
    
    train_precs.append(precision_score(y_train, y_train_pred))
    test_precs.append(precision_score(y_test, y_test_pred))
    
    train_recs.append(recall_score(y_train, y_train_pred))
    test_recs.append(recall_score(y_test, y_test_pred))
```

```python
plt.figure()

plt.subplots(1,2,figsize=(15,5))

plt.subplot(1,2,1)
plt.plot(np.arange(1,11,.5), train_accs, marker='.', label='Accuracy', color='skyblue')
plt.plot(np.arange(1,11,.5), train_precs, marker='.', label='Precision', color='orange')
plt.plot(np.arange(1,11,.5), train_recs, marker='.', label='Recall', color='salmon')
plt.xlabel('Weight for Class 1')
plt.ylabel('Score')
plt.xticks(range(1,11))
plt.legend()
plt.title('Train Data')


plt.subplot(1,2,2)
plt.plot(np.arange(1,11,.5), test_accs, marker='.', label='Accuracy', color='skyblue')
plt.plot(np.arange(1,11,.5), test_precs, marker='.', label='Precision', color='orange')
plt.plot(np.arange(1,11,.5), test_recs, marker='.', label='Recall', color='salmon')
plt.xlabel('Weight for Class 1')
plt.ylabel('Score')
plt.xticks(range(1,11))
plt.legend()
plt.title('Test Data')

plt.suptitle('Accuracy, Precision, and Recall for Models With Different Weights for Class 1', fontsize=14)
sns.despine()
plt.show()
```

```python
# Displaying the coefficients for the baseline logistic regression model

logreg_coef_dict = {}
for coef, feature in list(zip(logreg.coef_, X.columns)):
    logreg_coef_dict[feature] = coef

logreg_coef_df = pd.DataFrame(coef_dict, index=range(1))
logreg_coef_df.T.sort_values(by=[0], ascending=False)
```
